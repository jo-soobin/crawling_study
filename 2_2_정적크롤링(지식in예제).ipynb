```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import datetime as dt

url = 'https://kin.naver.com/search/list.naver?query=%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%90%98%EB%A9%B4'
html = requests.get(url)
soup = bs(html.text, 'html.parser')
df = pd.DataFrame(columns=['순서',"제목", "날짜"])
#s_content > div.section > ul
titles = soup.select('#s_content > div.section > ul')
title = titles[0].find_all('li')
num = len(title)
ranking=0

for i in range(num):
  date = pd.to_datetime(title[i].find('dd').text) 
  subject = (title[i].find('dt').text)
  ranking+=1
  df = df.append({'순서':ranking,'제목':subject.replace('\n',''),'날짜':date},ignore_index = True)
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
url = "https://kin.naver.com/search/list.naver?query=%EA%B2%B0%ED%98%BC%ED%95%98%EB%8A%94%EB%B0%A9%EB%B2%95"

html = requests.get(url)
soup = bs(html.text, "html.parser")

# selector 를 이용한 태그 검색
titles = soup.select("#s_content > div.section > ul")
title = titles[0].find_all("li")


df = pd.DataFrame(columns=["제목", "등록일","내용"])
for li in title:
  df = df.append({"등록일":li.find_all("dd")[0].text, "제목":li.find("dt").text.replace("\n",""),"내용":li.find_all("dd")[1].text}, ignore_index=True)
```
